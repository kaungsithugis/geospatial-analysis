```python
# import geopandas library
import geopandas as gpd
import matplotlib.pyplot as plt
```

GeoPandas can read almost any vector-based spatial data format including ESRI shapefile, GeoJSON files and more using the geopandas.read_file() command:

```python
# read geospatial data file
world_json = gpd.read_file("/content/drive/MyDrive/Data for Python/world_countries.json")

# exmine the data file
world_json.head()

# Basic plot, single color

world_json.plot()
```

```python
mm_shp = gpd.read_file("/content/drive/MyDrive/Data for Python/mmr_polbnda_250k_adm1_mimu.shp")

# exmine the data file
mm_shp.head()
```

```python
# Exmine Coordinate Reference system
mm_shp.crs
```

```python
# Basic plot, single color

mm_shp.plot();
```

Choropleth maps

GeoPandas makes it easy to create Choropleth maps (maps where the color of each shape is based on the value of an associated variable). Simply use the plot command with the column argument set to the column whose values you want used to assign colors.

```python
# Plot by state and region
mm_shp.plot(column="ST");
```

Plotting a map with legend

When plotting a map, one can enable a legend using the legend argument:

```python
# Plot state and region map with an accurate legend
mm_shp.plot(column='ST', figsize=(12, 12), edgecolor='black', legend=True, legend_kwds={'loc': 'lower left', 'fontsize': 9})
# Add title to the map
plt.title('Myanmar States and Regions Map', fontsize=16)
```
