```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
csv_file = '/content/drive/MyDrive/gis_data/eq_MM_last_20yrs.csv'
df = pd.read_csv(csv_file)
```

```python
latitude_column = 'latitude'
longitude_column = 'longitude'
magnitude_column = 'mag'

# Plot the earthquake data on a scatter plot
plt.figure(figsize=(8, 10))
plt.scatter(df[longitude_column], df[latitude_column], s=df[magnitude_column] * 100, c='red', alpha=0.1)
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Earthquake Locations')
plt.grid(True)
plt.show()
```
