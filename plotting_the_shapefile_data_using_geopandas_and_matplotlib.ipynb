```python
import geopandas as gpd
import folium
import matplotlib.pyplot as plt
```

```python
#read the shapefile using geopandas
shapefile_path = ("/content/drive/MyDrive/gis_data/gis_osm_railways_free_1.shp")
gdf = gpd.read_file(shapefile_path)
```

```python
# Plot the shapefile data using geopandas and matplotlib
figure, ax = plt.subplots(figsize=(10, 10))
gdf.plot(ax=ax)
```
