```python
import ee
import geemap
```

```python
ee.Authenticate()
```

```python
ee.Initialize(project='ee-kaungsithugeo')
```

```python
# Create an interactive map centered at specified coordinates and zoom level
Map = geemap.Map(center=[16,96], zoom=5)
Map
```

```python
# Import the USGS ground elevation image.
elv = ee.Image('USGS/SRTMGL1_003')
```

```python
# Define visualization parameters for the elevation layer
elevation_vis_params = {
    'min': 0,
    'max': 3000,
    'palette': ['0000FF', '00FFFF', '00FF00', 'FFFF00', 'FF0000', 'FFFFFF']
}
```

```python
# Add the elevation layer to the map using the defined visualization parameters
Map.addLayer(elv, elevation_vis_params, 'Elevation')
# Display the map
Map
```

```python
# Calculate the slope from the elevation data
slope = ee.Terrain.slope(elv)
```

```python
# Define visualization parameters for the slope layer
slope_vis_params = {
    'min': 0,
    'max': 60,
    'palette': ['0000FF', '00FF00', 'FFFF00', 'FFA500', 'FF4500', 'FF0000', '800000']
}
```

```python
# Add the slope layer to the map using the defined visualization parameters
Map.addLayer(slope, slope_vis_params, 'Slope')
```

```python
# Add a layer control panel to the map
Map.addLayerControl()

# Display the map
Map
```
